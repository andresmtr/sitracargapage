```python
# ============================================
# 1. Importar librer√≠as
# ============================================
import pandas as pd
import plotly.express as px

# ============================================
# 2. Cargar el archivo
#    Puedes usar CSV o Excel
# ============================================
# df = pd.read_csv("datos.csv")
df = pd.read_excel("/Users/andresmauriciotrianareina/Documents/GitHub/sitracargapage/assets/excel/Puntos_mapa.xlsx")

df.head()
```

```python
# ============================================
# 3. Separar la columna UNION en Lat y Lon
#    Ejemplo contenido: "4.6097, -74.0817"
# ============================================

# Separar por coma
df[["Latitud", "Longitud"]] = df["Union"].str.split(",", expand=True)

# Convertir a num√©ricos
df["Latitud"] = pd.to_numeric(df["Latitud"], errors="coerce")
df["Longitud"] = pd.to_numeric(df["Longitud"], errors="coerce")

df.head()
```

```python
# ============================================
# 4. Crear mapa de puntos con Plotly y guardar en HTML
# ============================================

fig = px.scatter_mapbox(
    df,
    lat="Latitud",
    lon="Longitud",
    hover_name="Direccion",
    hover_data={
        "Departamento": True,
        "Ciudad": True,
        "Telefono": True,
        "Latitud": False,     # se ocultan
        "Longitud": False
    },
    zoom=4.9,
    height=600
)

# Usar mapa gratuito de OpenStreetMap
fig.update_layout(mapbox_style="open-street-map")

# Reducir m√°rgenes
fig.update_layout(margin={"r":0, "t":0, "l":0, "b":0})

# Tama√±o de los puntos
fig.update_traces(marker=dict(size=14))

# Tooltip personalizado
fig.update_traces(
    hovertemplate=
        "<b>Direcci√≥n:</b> %{hovertext}<br>" +
        "<b>Departamento:</b> %{customdata[0]}<br>" +
        "<b>Ciudad:</b> %{customdata[1]}<br>" +
        "<b>Tel√©fono:</b> %{customdata[2]}<br>" +
        "<extra></extra>"
)

# üîπ Mostrar en el notebook
fig.show()

# üîπ Guardar como archivo HTML interactivo
fig.write_html("mapa_puntos.html")
```
